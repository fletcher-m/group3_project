# Extracting the Title of a Wikipedia Page:

```````python
# Replace this with the URL of the webpage you want to scrape
url = 'https://en.wikipedia.org/wiki/2020_United_States_census'


# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the title tag and print its text
    title_tag = soup.title
    if title_tag:
        print('Title:', title_tag.text)
    else:
        print('No title found on the page.')
else:
    print('Failed to retrieve the web page.')








```````

# Get the Anchor Tags From a Webpage

```````python
# Replace 'url' with the URL of the webpage you want to scrape
url = 'https://en.wikipedia.org/wiki/2020_United_States_census'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # The HTML content of the webpage is in response.text
    html_content = response.text

    # Create a BeautifulSoup object
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract all links (anchor tags) from the webpage
    links = soup.find_all('a')
    
    # Print the href attribute of each link
    for link in links:
        print(link.get('href'))
else:
    print('Failed to retrieve the web page.')
```````

```````python
# Define the URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/Vermont'

# Send an HTTP GET request to the URL
response = requests.get(url)

if response.status_code == 200:
    html_content = response.text
else:
    print('Failed to retrieve the Wikipedia page.')
    exit()

# Create a Beautiful Soup object
soup = BeautifulSoup(html_content, 'html.parser')

# Find all tables in the content
tables = soup.find_all('table')

# Check if the fourth table exists (Population by Race table)
if len(tables) > 3:
    table = tables[3]
else:
    print('Population by Race table not found on the page.')
    exit()

# Parse and process the table data
if table:
    rows = table.find_all('tr')
    for row in rows:
        cells = row.find_all(['td', 'th'])
        row_data = [cell.get_text(strip=True) for cell in cells]
        print(row_data)
```````

```````python

# Parse the table data and store it in a list of lists
table_data = []
if table:
    rows = table.find_all('tr')
    for row in rows:
        cells = row.find_all(['td', 'th'])
        row_data = [cell.get_text(strip=True) for cell in cells]
        table_data.append(row_data)

# Convert the list of lists into a Pandas DataFrame
df = pd.DataFrame(table_data[1:], columns=table_data[0])

# Print the DataFrame
print(df)
```````

```````python
plt.figure(figsize=(10, 6))
plt.bar(df['Town'], df['Population'], color='skyblue')
plt.ylabel('Population')
plt.title('Population of Vermont by Town')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.gca().invert_yaxis()


# Display the chart
plt.show()
```````
